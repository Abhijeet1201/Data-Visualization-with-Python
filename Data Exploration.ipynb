```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('https://raw.githubusercontent.com/ChrisWalshaw/DataViz/master/Data/001129838/DailyCustomers.csv', index_col=0)
data.index = pd.to_datetime(data.index)
```

Here the data to be explored has been imported

```python
print(data.head())
print(data.tail())
print(data.describe())
print(data.sum())
print(data.mean())
```

Basic explorations used on the data to get more insights. Here Sum shows the total number of customers who visited that particular store in 2019 and mean shows the average of customers who visited that particular store in 2019

```python
data = data.reindex(data.mean().sort_values(ascending=False).index, axis=1)
print(data.mean())
print(data.head())
print(data.tail())

```

Here data has been sorted in ascending order to show the highest performing stores in terms of average of the total number of customers visiting that store. It can be seen that RAH, SGA, SMM, QSN, PAA are the top 5 highest performing stores in terms of total clients visiting that stor and ZMS, NGB, YYO, MAJ, ZSD are the worst performing stores.


```python
selected = ['RAH', 'SGA', 'SMM', 'QSN', 'PAA']

data = data[selected]
print(data.head())
print(data.tail())
data.plot.line()

plt.show()
```

The above visualization shows line graph of top 5 most visited stores. It can be seen that RAH had the highest amount of daily customers in mid-may.

```python
selected = ['ZMS', 'NGB', 'YYO', 'MAJ', 'ZSD']

data = data[selected]
print(data.head())
print(data.tail())
data.plot.line()

plt.show()
```

The above visualization shows line graph of bottom 5 most stores. 

```python
data2 = data[selected].sum()
print(data2.head())
data2.plot.bar()
plt.show()
```

The above bar chart shows the total number of customers who visited in the top 5 performing stores.

```python
data = data.loc[pd.to_datetime('2019-10-01'): pd.to_datetime('2019-12-31')]
print(data.head())
print(data.tail())

data.plot.line()
plt.show()
```
